## f) Сгенерируйте выборку объема n = 100 для некоторого значения параметра $\theta$. Вычислите указанные выше доверительные интервалы для доверительной вероятности 0.95






```python
theta = 5

import numpy as np
import random

def p(x):
    return 1 / theta if theta <= x <= 2 * theta else 0

def F(x):
    if (x <= theta):
        return 0
    elif x >= 2 * theta:
        return 1
    else:
        return (x / theta - 1)

def F_inv(y):
    return theta * (y + 1)
    
n = 100
array= [float(F_inv(random.random())) for _ in range(n)]

print(array)
```

### Точный метод:

$$
P(\frac{max(x_i)}{\sqrt[n]{\frac{1 + \beta}{2}} + 1} < \theta < \frac{max(x_i)}{\sqrt[n]{\frac{1 - \beta}{2}} + 1}) = \beta
$$

### Асимптотический метод:

$$
P(\tilde{\theta}_1 - \frac{t_2 \cdot 2S\sqrt{n - 1}}{3n} < \theta < \tilde{\theta}_1 - \frac{t_1 \cdot 2S\sqrt{n - 1}}{3n}) = \beta
$$

```python
print("\tТОЧНЫЙ")

beta = 0.95
left = np.max(array) / (((1 + beta) / 2) ** (1 / n) + 1)
right = np.max(array) / (((1 - beta) / 2) ** (1 / n) + 1)

print(f'Доверительный интервал: ({left}, {right})')
print(f'Длина интервала: {right - left}')


print("\n\tАСИМПТОТИЧЕСКИЙ")

t_1 = -1.96
t_2 = 1.96

theta_1 = 2 / 3 * np.mean(array)
s = np.sqrt(np.sum([(x - np.mean(array)) ** 2 for x in array]) / (n - 1))

left = theta_1 - (t_2 * 2 * s * np.sqrt(n - 1)) / (3 * n)
right = theta_1 - (t_1 * 2 * s * np.sqrt(n - 1)) / (3 * n)

print(f'Доверительный интервал: ({left}, {right})')
print(f'Длина интервала: {right - left}')
```

## g) Численно постройте бутстраповский доверительный интервал.

```python
bootstrap_iteration = 1000

theta_with_wave = (n + 1) / (2 * n + 1) * np.max(array) 

bootstrap_delta = []

for i in range(bootstrap_iteration):
    bootstrap_array = np.random.choice(array, size=len(array), replace=True)
    bootstrap_theta = (n + 1) / (2 * n + 1) * np.max(bootstrap_array)
    bootstrap_delta.append(bootstrap_theta - theta_with_wave)

bootstrap_delta.sort()
t_1 = bootstrap_delta[int(bootstrap_iteration * (1 - beta) / 2)]
t_2 = bootstrap_delta[int(bootstrap_iteration * (1 + beta) / 2)]

right = -(t_1 - theta_with_wave)
left = -(t_2 - theta_with_wave)

print(f'Доверительный интервал: ({left}, {right})')
print(f'Длина интервала: {right - left}')
```
