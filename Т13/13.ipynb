```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import f

n, m = 139, 1000

s1_l, s1_w = 5.722, 4.612
s2_l, s2_w = 6.161, 5.055

alpha = 0.05

f_low, f_high = f.ppf(alpha/2, n-1, m-1), f.ppf(1-alpha/2, n-1, m-1)

f_len, f_wid = s1_l**2 / s2_l**2, s1_w**2 / s2_w**2

print(f"F-стат ширины: {f_wid:.4f}")
print(f"F-стат длины: {f_len:.4f}")
print(f"Gкр: <={f_low:.4f} и >={f_high:.4f}")
```

$F$длины и $F$ширины $\notin$ Gкр $\implies$ Нет оснований отвергнуть H0

```python
plt.figure(figsize=(8, 4))

sigma_ratio = np.linspace(0.5, 2, 100)  

power = 1 - f.cdf(f_high, n-1, m-1, scale=sigma_ratio) + f.cdf(f_low, n-1, m-1, scale=sigma_ratio)

plt.axvline(x=f_len, color='pink', linestyle=':', alpha=0.7, label=f'Длина (x={f_len:.2f})')
plt.axvline(x=f_wid, color='mediumvioletred', linestyle=':', alpha=0.7, label=f'Ширина (x={f_wid:.2f})')

plt.plot(sigma_ratio, power, 'r-', linewidth=2, label='Мощность критерия', color='palevioletred')

plt.title('Мощность критерия')
plt.xlabel('σ1²/σ2²')
plt.ylabel('W')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(0, 1.01)

plt.show()
```
