```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from math import sqrt
```

```python
x, y = np.array([-1.11, -6.1, 2.42]), np.array([-2.29, -2.91])

n, m = 3, 2

s_x, s_y = 2, 1

alpha = 0.05

delta = (x.mean() - y.mean()) / sqrt(s_x / n + s_y / m)

p_val = norm.sf(delta, 0, 1)
print(f" p-val = {p_val:.4f}")
```

p-val > $\alpha$ $\implies$ Нет оснований отвергнуть H0

График мощности критерия:

```python

plt.figure(figsize=(10, 6))

effect_sizes = np.linspace(0, 5, 200)

power = []
for effect in effect_sizes:
    z_alpha = norm.ppf(1 - alpha)
    mean_alt = effect / sqrt(s_x/n + s_y/m)
    power.append(norm.sf(z_alpha - mean_alt))

plt.plot(effect_sizes, power, 'b-', linewidth=2, label='Мощность критерия', color='palevioletred')

plt.axvline(x=0, color='mediumvioletred', linestyle='--', 
    label=f'H0, a - b = 0')

plt.title('Мощность')
plt.xlabel('a - b')
plt.ylabel('W')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(0, 1.01)

plt.show()
```
